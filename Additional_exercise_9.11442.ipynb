```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
label=pd.DataFrame(list(map(lambda x: iris.target_names[x],iris.target)),columns=['Species'])
iris=pd.concat([data,label],axis=1)
print(iris.head())
```

# 1.Use the distplot() to see the distribution of  the SepalLengthCm, SepalWidthCm, PetallengthCm,PetalwidthCm. Plot them as subplots in a single image

```python
sns.set_style("whitegrid");
fig, axs = plt.subplots(figsize=(6,6), ncols=2, nrows=2)
sns.distplot(iris['sepal length (cm)'],ax=axs[0, 0])
sns.distplot(iris['sepal width (cm)'], ax=axs[0, 1])
sns.distplot(iris['petal length (cm)'], ax=axs[1, 0])
sns.distplot(iris['petal width (cm)'], ax=axs[1, 1])
plt.show();
```

# 3)Do a countplot for the feature 'Species'

```python
sns.countplot(x='Species',data=iris,palette="Set1")
```

# 2) Do a boxplot of all the features except  'Species"

```python
fig, axes = plt.subplots(2,2)
for i,el in enumerate(list(iris.columns.values)[:-1]):
    a = iris.boxplot(el, ax=axes.flatten()[i])

plt.show()
```

# 4.Do a pairplot on the features  'Sepallengthcm',' Sepalwidthcm','PetalLengthcm','Petalwidthcm','Species'

```python
sns.set_style("darkgrid");
sns.pairplot(iris, hue="Species", height=3);
```

# 6. Do a bar plot of Species vs sepal length (cm)

```python
sns.barplot(x='Species',y='sepal length (cm)',data=iris)
```

# 5. Do an lmplot on the following SepalLengthcm, PetalLengthcm. Using hue, display the different species in different colours

```python
sns.lmplot(x='sepal length (cm)',y='petal length (cm)',data=iris,hue='Species')
```

# 7. Using heatmap, plot the correlation matrix calculate the correlation matrix


```python
sns.heatmap(iris.corr(),cmap="YlGnBu", linecolor='white', linewidths=1)
```
